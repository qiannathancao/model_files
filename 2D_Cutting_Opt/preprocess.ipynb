```
def col_name(df):
    """
    this is to trim the data_frame column names to a unique format:
    all case, replace space to underscore, remove parentheses
    param df:
        raw from share drive for
    return:
        polished data set with new column names
    """
    res = [c.strip().lower().replace('-', '').replace(' ', '_').replace('(', '').replace(')', '').replace('"', '') for c in df.columns]
    return res
```

```
%sql
drop table if exists gpsc.fabrication_cutting_yield_comm;
create table gpsc.fabrication_cutting_yield_comm
using csv
LOCATION '/mnt/ml/com/dmnd00xxxxx_fabrication_cutting_yield_optimization/input_data/fabrication_material_sdx_comm_04132020.csv'
OPTIONS(header 'true');
```

```
df = spark.table('gpsc.fabrication_cutting_yield_comm')
col_list = col_name(df)
df_clean = df.toDF(*col_list)
```
