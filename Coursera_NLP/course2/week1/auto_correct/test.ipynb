```python
word = 'something'
```

```python
letters = 'abcdefghijklmnopqrstuvwxyz'
```

```python
split = [(word[:i], word[i:]) for i in range(len(word)+1)]
```

```python
split
```

```python
delete = [L + R[1:] for L, R in split if R ]
delete
```

```python
transpose = [L + R[1] + R[0] + R[2:] for L, R in split if len(R) > 1]
transpose
```

```python
replace = [L + c + R[1:] for L , R in split if R  for c in letters]
replace
```

```python
inserts = [L+ c + R for L, R in split for c in letters]
inserts
```

```python
ret = set(delete+transpose+replace+inserts)
```
